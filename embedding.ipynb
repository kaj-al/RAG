```python
from dotenv import load_dotenv
load_dotenv()
```

```python

from langchain_huggingface import HuggingFaceEmbeddings

docs = ["Artificial Intelligence = machines simulating human intelligence",
"Learns from data and experience",
'Performs reasoning and decision-making',
"Solves problems automatically",
"Understands and processes language",
"Recognizes patterns and images",
'Includes Machine Learning and Deep Learning',
"Used in assistants, recommendations, self-driving cars",
"Aims to reduce human effort and increase efficiency" ]

embeddings = HuggingFaceEmbeddings(model="all-MiniLM-L6-v2")

vector = embeddings.embed_documents(docs)
len(vector[0])
 
```

```python
from langchain_chroma import Chroma

store = Chroma.from_texts(
    texts = docs,
    embedding = embeddings,
    persist_directory = "./vector_db" #to store it locally
)
```

```python
query = "What is AI?"
ans = store.similarity_search(query=query,k=2)
ans
```
