Text Splitting

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

texts = "Natural environment, refers to all living and non-living things occurring naturally and the physical and biological factors along with their chemical interactions that impact on any organism or a group of organisms. "
splitter = RecursiveCharacterTextSplitter(chunk_size=200,chunk_overlap=100)
chunks = splitter.split_text(texts)
chunks[0]
len(chunks)  
```

Document Splitting

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

loader = PyPDFLoader("nexora.pdf")
docs = loader.load()
splitter = RecursiveCharacterTextSplitter(chunk_size=200,chunk_overlap=100)
chunks = splitter.split_documents(docs)
chunks[0]
len(chunks)  
```
