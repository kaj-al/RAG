Text Loader

```python

from langchain_community.document_loaders import TextLoader

loader = TextLoader("text.txt")
text = loader.load()
text[0].page_content 
len(text)
```

PDF Loader

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("nexora.pdf")
pdf = loader.load()
pdf[0].page_content 
len(pdf)
```

CSV Loader

```python
from langchain_community.document_loaders import CSVLoader

loader = CSVLoader("city_day.csv")
csv = loader.load()
csv[0].page_content 
len(csv)
```

Web Page Loader

```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://docs.langchain.com/")
web = loader.load()
web[0].page_content 
len(web)
```

Wikipedia Loader

```python
from langchain_community.document_loaders import WikipediaLoader

loader = WikipediaLoader("https://en.wikipedia.org/wiki/Environment",load_max_docs=2)
wiki = loader.load()
wiki[0].page_content 
len(wiki)
```
